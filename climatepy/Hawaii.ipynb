```python
# import dependencies
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')
import pandas as pd
import datetime as dt
import numpy as np
from scipy import stats
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func, desc
```

```python
engine = create_engine("sqlite:///./Resources/hawaii.sqlite", echo=False)
```

```python
# Reflect Database into ORM classes
Base = automap_base()
Base.prepare(engine, reflect=True)
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Map Measurement and station classes
measurement = Base.classes.measurement
station = Base.classes.station
```

```python
# create a session
session = Session(engine)
```

```python
# Get a list of column names and types of measurement table
inspector = inspect(engine)
columns = inspector.get_columns('measurement')

for c in columns:
    print(c["name"], c["type"])
```

```python
# Get a list of column names and types of station table
columns = inspector.get_columns('station')

for c in columns:
    print(c["name"], c["type"])
```

# Precipitation Analysis
- - -

```python
# Get the latest date in the database to determine the date to use for the query (1 year before last date)
latest_date=session.query(measurement.date).order_by(measurement.date.desc()).first()[0]
latest_date_object=dt.datetime.strptime(latest_date,'%Y-%m-%d')
query_date = latest_date_object - dt.timedelta(days=365)
query_date
```

```python
# Perform a query to retrieve the data and precipitation scores
sel = [measurement.date, measurement.prcp]
precipitation=session.query(*sel).\
    filter(measurement.date>query_date).all()
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
precipitation_df = pd.DataFrame(precipitation, 
        columns=["date","precipitation"])
precipitation_df.set_index("date", inplace=True)
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
precipitation_df.plot(rot=90,
                      use_index=True,
                      title='Last Twelve Months Precipitation in Hawai',
                      grid=True,
                      legend=False,
                      figsize=(15,6)
)
plt.ylabel('Precipitation')
plt.savefig('./images/precipitation.png')
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
precipitation_df.describe()
```

# Station Analysis
- - -

```python
# Design a query to show how many stations are available in this dataset?
sel = [station.station, station.name]
print(f"There are {session.query(*sel).count()} measurement stations")
```

```python
# Query the total measurements (temperature and precipitation) in order to sort the stations based on activity
sel = [station.station, station.name,func.count(measurement.prcp)+func.count(measurement.tobs)]
stations_activity=session.query(*sel).filter(station.station==measurement.station).\
group_by(measurement.station).order_by(func.count(measurement.tobs).desc()).all()
stations_activity
```

```python
#Which station has the highest number of observations?
most_active_id,most_active_name,most_active_measurements=stations_activity[0]
print(f"The most active station was '{most_active_name}' with {most_active_measurements} total observations")
```

```python
#Design a query to retrieve the last 12 months of temperature observation data (tobs).
sel=[measurement.tobs]
temperature_data=session.query(*sel).filter(station.station==measurement.station).\
filter(station.station==most_active_id).filter(measurement.date>query_date).all()
#convert the results into an array
tempearature_array=[]
for t in temperature_data:
    tempearature_array.append(t[0])
```

```python
#Plot the results as a histogram with bins=12.
plt.hist(tempearature_array,bins=12,edgecolor='white')
plt.xlim(min(tempearature_array),max(tempearature_array))
plt.title("Observed temperatures in Waihee station")
plt.ylabel("Frequency")
plt.xlabel("Temperature (°F)")
plt.savefig('./images/temperature_histogram.png')
plt.show
```

```python
#Identify the average temperature in June at all stations across all available years in the dataset. 
june_temps=session.query(measurement.tobs).filter(func.strftime("%m", measurement.date)=="06").all()
june_array=[]
for t in june_temps:
    june_array.append(t[0])    
june_avg=sum(june_array)/len(june_array)
#Do the same for December temperature.
dec_temps=session.query(measurement.tobs).filter(func.strftime("%m", measurement.date)=="12").all()
dec_array=[]
for t in dec_temps:
    dec_array.append(t[0])
dec_avg=sum(dec_array)/len(dec_array)
print(f"The average temperature in June is {round(june_avg,2)} and in December is {round(dec_avg,2)}")
```

```python
#Use the t-test to determine whether the difference in the means, if any, is statistically significant
stats.ttest_ind(june_array,dec_array,equal_var = False)
# Result indicates that the difference in the means in significant
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(measurement.tobs), func.avg(measurement.tobs), func.max(measurement.tobs)).\
        filter(measurement.date >= start_date).filter(measurement.date <= end_date).all()
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.
temp_stats=calc_temps("2017-01-01","2018-01-01")[0]
temp_stats
```
